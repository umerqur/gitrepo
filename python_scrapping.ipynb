```python
#Use Webscraping to Extract Tesla Revenue Data
#Display the last five rows of the tesla_revenue dataframe using the tail function.

import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://en.wikipedia.org/wiki/Tesla,_Inc.#Finances'

# Fetch the HTML content of the webpage
response = requests.get(url)
html_content = response.text

# Use BeautifulSoup to parse the HTML content

soup = BeautifulSoup(html_content, 'html.parser')

tables_with_class = soup.find_all('table', {'class': 'wikitable'})

specific_table = tables_with_class[1]  # second table with this class name
    
# Extract the data into a DataFrame
table_data = pd.read_html(str(specific_table))
    
# Access the DataFrame and display or process the data
specific_table_df = table_data[0]
    
#Select last five rows using tail function
last_five_rows = specific_table_df.tail()

selected_columns = ['Year', 'Revenue (US$ m)']

# Subsetting the DataFrame to include only the selected columns
specific_table_subset = specific_table_df[selected_columns]

# Getting the last five rows of the subset DataFrame
last_five_rows_subset = specific_table_subset.tail()
print(last_five_rows_subset)

```

```python
import yfinance as yf
import matplotlib.pyplot as plt

def make_graph(stock_data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data['Close'], label='Tesla Stock Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.grid(True)
    plt.show()

# Fetch Tesla stock data from Yahoo Finance
ticker = 'TSLA'
stock = yf.Ticker(ticker)
data = stock.history(period='max')

# Call make_graph function to plot the data with a title
make_graph(data, 'Tesla Stock Price (Max History)')
```
